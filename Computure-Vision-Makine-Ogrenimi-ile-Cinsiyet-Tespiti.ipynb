```python
import cv2
import matplotlib.pyplot as plt
import cvlib
from cvlib.object_detection import draw_bbox
```

```python
img = cv2.imread('foto3.jpg')
```

```python
plt.imshow(img)
```

```python
bbox, label, conf = cvlib.detect_common_objects(img)
```

```python
output = draw_bbox(img, bbox, label, conf)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(output)
```
